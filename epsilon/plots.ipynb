```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import entropy
import itertools
```

```python
decays = [1000, 2000, 5000, 7000, 10000]
episodes = []
scores = []
for decay in decays:
    episodes.append(np.load(f"{decay}_episodes.npy"))
    scores.append(np.load(f"{decay}_score.npy"))
```

```python
for i, decay in enumerate(decays):
    plt.plot(scores[i], label=f"decay={decay}")
    print(scores[i].shape)
    plt.legend()
    plt.show()
```

```python
steps = np.linspace(0, 100000, len(episodes[0]))
for i, decay in enumerate(decays):
    plt.plot(steps, episodes[i], label=f"decay={decay}")
plt.legend()
plt.show()
```

```python
coverage_steps = [200, 500, 1000, 50000]
heat = []
for decay in decays:
    for step in coverage_steps:
        heat.append(np.load(f"{decay}_heat_{step}.npy"))
```

```python
entropies = []
for i in range(len(decays)*len(coverage_steps)):
    plt.subplot(len(decays), len(coverage_steps), i+1)
    plt.imshow(heat[i], cmap='hot', interpolation='nearest')
    plt.axis('off')
    entropies.append(entropy(heat[i].ravel()))
plt.show()
```

```python
print(entropies)
entropies = np.array(entropies).reshape(len(decays), len(coverage_steps))
print(entropies)

    
```
